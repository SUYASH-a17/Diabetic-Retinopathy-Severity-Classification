- Importing required libraries

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import split, col
from pyspark.sql.types import DoubleType
```

```python
# Initialized Spark Session
spark = SparkSession.builder.getOrCreate()
```

```python
raw_df = spark.read.option("header", "false").csv("dbfs:/FileStore/tables/train_features_dr_unified_og.csv")
raw_df.show(5) 
```

- We detect if the DataFrame has all features in one column and, if so, split it into 2047 individual feature columns plus a label column to structure the data for analysis and modeling.

```python
if len(raw_df.columns) == 1:
    split_cols = split(col("_c0"), ",")
    num_cols = 2048  # 2047 features + 1 label

    new_df = raw_df.select(
        [split_cols.getItem(i).cast(DoubleType()).alias(f"f{i}") for i in range(num_cols - 1)] +
        [split_cols.getItem(num_cols - 1).cast(DoubleType()).alias("label")]
    )

    display(new_df.limit(15))  # Getting preview 
else:
    print("Unexpected format: already has multiple columns. Inspect manually.")
```

- Write the cleaned, structured feature DataFrame back to DBFS as partitioned CSV files for downstream processing and easy access.  

```python
# Save to DBFS (as a folder with part files)
new_df.write.mode("overwrite").option("header", "true").csv("dbfs:/FileStore/tables/clean_train_features")

print("Cleaned CSV saved to: dbfs:/FileStore/tables/clean_train_features")
```

- We save a 50-row sample of the cleaned training features to DBFS so we can quickly preview and validate the preprocessing results before full-scale processing.

```python
new_df.limit(50).write.mode("overwrite").option("header", "true").csv("dbfs:/FileStore/tables/sample_clean_train_features")
```

- Splitting the cleaned DataFrame into manageable batches improves write performance and avoids timeouts or memory issues when saving large CSV files to DBFS

```python
total_rows = new_df.count()
batch_size = 500

print(f"Total rows: {total_rows} | Writing in batches of {batch_size} rows...\n")

for i in range(0, total_rows, batch_size):
    start = i
    end = min(i + batch_size, total_rows)
    
    # Slice using subtract-based trick (since Spark DataFrames are not indexed)
    batch_df = new_df.limit(end).subtract(new_df.limit(start))
    
    # Define unique output path for each batch
    output_path = f"dbfs:/FileStore/tables/clean_train_features_batch_{start}_{end - 1}"
    
    # Write batch
    batch_df.write.mode("overwrite").option("header", "true").csv(output_path)
    
    print(f"Saved rows {start} to {end - 1}  {output_path}")
```

- We repartition the DataFrame for optimized parallelism and then write the fully cleaned training features to DBFS as the final dataset for downstream processing.

```python
new_df.repartition(10).write.mode("overwrite").option("header", "true").csv("dbfs:/FileStore/tables/clean_train_features_final")
```
